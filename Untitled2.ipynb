```python
import numpy as np

np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X.reshape(-1, 1), y)
y_pred = lr.predict(X.reshape(-1, 1))
```

```python
plt.scatter(X, y, color='b', label='X')
plt.plot(X, y_pred, color='r', label='y')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0
        
        for _ in range(self.num_iterations):
            y_pred = self.predict(X)
            
            # 경사 하강법 업데이트
            dw = (1 / num_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / num_samples) * np.sum(y_pred - y)
            
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

# __init__(self, learning_rate=0.01, num_iterations=1000): 클래스의 초기화 메서드입니다. 
# 이 메서드는 학습률(learning_rate)과 반복 횟수(num_iterations)를 인자로 받습니다. 
# 초기화 과정에서는 학습률과 반복 횟수를 멤버 변수로 저장하고, 가중치(weights)와 편향(bias)을 초기화합니다.

# fit(self, X, y): 주어진 입력 데이터 X와 목표 변수 y를 사용하여 모델을 학습합니다. 
# 이 메서드에서는 경사 하강법을 사용하여 가중치와 편향을 업데이트합니다. 
# 학습 데이터의 특징 개수와 샘플 개수를 가져와 가중치와 편향을 0으로 초기화한 후, 지정된 반복 횟수만큼 반복하면서 예측 값을 계산하고 가중치와 편향을 업데이트합니다.

# predict(self, X): 주어진 입력 데이터 X를 사용하여 예측 값을 계산합니다. 
# 이 메서드는 입력 데이터와 가중치, 편향을 사용하여 선형 회귀 모델의 예측 값을 계산합니다. 
# 넘파이의 dot 함수를 사용하여 입력 데이터와 가중치의 내적을 계산하고, 이에 편향을 더해 예측 값을 얻습니다.


# 예시 데이터
np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)

# LinearRegression 모델 인스턴스 생성 및 학습
model = LinearRegression(learning_rate=0.01, num_iterations=1000)
model.fit(X.reshape(-1, 1), y)

# 예측
X_test = np.linspace(0, 5, 100)
y_pred = model.predict(X_test.reshape(-1, 1))

# 산점도 그래프
plt.scatter(X, y, color='blue', label='Actual')

# 선 그래프
plt.plot(X_test, y_pred, color='red', label='Predicted')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

```python

class LinearRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0
        
        for _ in range(self.num_iterations):
            y_pred = self.predict(X)
            
            # 경사 하강법 업데이트
            dw = (1 / num_samples) * np.dot(X.T, (y_pred - y))
            db = (1 / num_samples) * np.sum(y_pred - y)
            
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
    
    def get_sum_of_coefficients(self):
        return np.sum(self.weights) + self.bias
    
    def partial_derivative_weights(self, X, y):
        num_samples = X.shape[0]
        y_pred = self.predict(X)
        return (1 / num_samples) * np.dot(X.T, (y_pred - y))
    
    def partial_derivative_bias(self, X, y):
        num_samples = X.shape[0]
        y_pred = self.predict(X)
        return (1 / num_samples) * np.sum(y_pred - y)

np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)

model = LinearRegression(learning_rate=0.01, num_iterations=1000)
model.fit(X.reshape(-1, 1), y)

X_partial_weights = model.partial_derivative_weights(X.reshape(-1, 1), y)
bias_partial = model.partial_derivative_bias(X.reshape(-1, 1), y)

print("Partial derivative w.r.t. weights:", X_partial_weights)
print("Partial derivative w.r.t. bias:", bias_partial)
```
